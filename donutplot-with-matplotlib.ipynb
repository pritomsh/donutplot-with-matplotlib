```python
# first import library
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
data = pd.read_csv('data/Customer-Churn-Prediction.csv')
data.head()
```

```python
# here groupby Payment Method and make a new data 
newdata = data.groupby('gender').size().rename_axis('gender').reset_index(name='Count')
newdata.head()
```

```python
# make 2 variable for labels and values 
labels = newdata.gender
values = newdata.Count
```

# Visualization

###  In matplotlib we can't directly call donutplot. There is a way to make donutplot.
    - First we have to make a pic plot 
    - than make a circle and change circle color same as         backgound color
   

```python
# Create a pieplot
plt.pie(values,)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

#show plot
plt.show()
```

```python
# add labels 
```

```python
# Create a pieplot
plt.pie(values, labels=labels,)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

#show plot
plt.show()
```

```python
# You can customize the distances of labels from the chart by using labeldistance parameter in the pie() function

 
# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')

# Label distance: gives the space between labels and the center of the pie
plt.pie(values, labels=labels, labeldistance=0.45)
p = plt.gcf()
p.gca().add_artist(my_circle)

#show plot
plt.show()
```

# Give Legend 

```python
# Create a pieplot


plt.pie(values, labels=labels,)
plt.legend(loc='upper left')


# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

#show plot
plt.show()
```

# change labels color

```python
# It is possible to change the colors of labels with rcParams:


# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')
 
# Label color
plt.rcParams['text.color'] = 'red'
plt.pie(values, labels=labels,)
p = plt.gcf()
p.gca().add_artist(my_circle)


#show plot
plt.show()
```

# add title

```python
# It is possible to change the colors of labels with rcParams:


# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')
 
plt.title('This is Donutplot', color='black');    
# Label color
plt.rcParams['text.color'] = 'red'
plt.pie(values, labels=labels,)
p = plt.gcf()
p.gca().add_artist(my_circle)


#show plot
plt.show()
```

# Wedges
wedgeprops parameter can be passed to the pie() function in order to set width of the wedge border lines and color:

```python
# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')

# Custom wedges
plt.pie(values, labels=labels, wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

# Changing the background color of a donut plot

```python
'''In order to change the background color, you should create a figure with figure() function and set the color with patch.
set_facecolor() before creating your pie plot.'''


 
# create a figure and set different background
fig = plt.figure()
fig.patch.set_facecolor('black')
 
# Change color of text
plt.rcParams['text.color'] = 'white'
 
# Create a circle at the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='black')
 
# Pieplot + circle on it
plt.pie(values, labels=labels,)
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
